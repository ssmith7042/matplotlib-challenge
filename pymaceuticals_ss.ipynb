### Analysis

As shown in the summary statistics table and visualized by the box plots below, Capomulin and Ramicane were the most effective treatments, achieving the lowest tumor volumes.  The final tumor volume for these treatment groups showed similar outcomes, with Ramicane having slightly better results.  The summary statistics table suggests that all of the other treatment regimens performed similarly and notably worse than Capomulin and Ramicane based on the measure of their groups' tumor volumes.  

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from scipy.stats import linregress
```

```python
# Study data files
mouse_metadata_path = "Mouse_metadata.csv"
study_results_path = "Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame

combined_df=mouse_metadata.merge(study_results, how='left', on='Mouse ID')
combined_df
# Display the data table for preview
```

```python
# Checking the number of mice.
NumberOfMice=len(pd.unique(combined_df['Mouse ID']))
NumberOfMice
len(combined_df)
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
combined_df.loc[combined_df.duplicated(['Mouse ID','Timepoint'], keep=False)==True]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned=combined_df[combined_df['Mouse ID']!='g989']
cleaned
```

```python
# Checking the number of mice in the clean DataFrame.
len(cleaned)
```

## Summary Statistics

```python
summary_statistics=cleaned.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].agg(['var','median', 'mean','std','sem'])
summary_statistics
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_measure=cleaned.groupby('Mouse ID').max('Timepoint')['Timepoint']
last_measure=last_measure.reset_index()
last_measure
```

```python
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merge_last=last_measure.merge(cleaned, on=['Mouse ID','Timepoint'],how='left')
merge_last
```

```python
# Put treatments into a list for for loop (and later for plot labels): Capomulin, Ramicane, Infubinol, and Ceftamin
treatments=['Capomulin','Ramicane','Infubinol','Ceftamin']
# Create empty list to fill with tumor vol data (for plotting)
tumor_vols=[]

for treatment in treatments:
    
    last_tv=merge_last.loc[merge_last['Drug Regimen']==treatment]['Tumor Volume (mm3)']
    quartiles=last_tv.quantile([.25,.5,.75])
    lowerq=quartiles[0.25]
    upperq=quartiles[0.75]
    iqr=upperq-lowerq
    lower_bound=lowerq-(1.5*iqr)
    upper_bound=upperq+(1.5*iqr)
    outliers=[]
    for measure in last_tv:
        if measure<lower_bound or measure>upper_bound:
            outliers.append(measure)
    print(f"{treatment} potential outliers are {outliers}")
    print(f"{treatment} interquartile range is {iqr}")

    
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
capomulin=merge_last.loc[(merge_last['Drug Regimen']=='Capomulin'),:]
ramicane=merge_last.loc[(merge_last['Drug Regimen']=='Ramicane'),:]
infubinol=merge_last.loc[(merge_last['Drug Regimen']=='Infubinol'),:]
ceftamin=merge_last.loc[(merge_last['Drug Regimen']=='Ceftamin'),:]

fig1, ax1 = plt.subplots()
ax1.set_title('Capomulin Outcomes')
ax1.set_ylabel('Tumor Volume (mm3)')
ax1.boxplot(capomulin['Tumor Volume (mm3)'])

fig2, ax2 = plt.subplots()
ax2.set_title('Ramicane Outcomes')
ax2.set_ylabel('Tumor Volume (mm3)')
ax2.boxplot(ramicane['Tumor Volume (mm3)'])

fig3, ax3 = plt.subplots()
ax3.set_title('Infubinol Outcomes')
ax3.set_ylabel('Tumor Volume (mm3)')
ax3.boxplot(infubinol['Tumor Volume (mm3)'])

fig4, ax4 = plt.subplots()
ax4.set_title('Ceftamin Outcomes')
ax4.set_ylabel('Tumor Volume (mm3)')
ax4.boxplot(ceftamin['Tumor Volume (mm3)'])
plt.show()

```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
isolate_957=cleaned.loc[cleaned['Mouse ID']=='m957']
plt.plot(isolate_957.Timepoint.tolist(),isolate_957['Tumor Volume (mm3)'], color='red', label='m957')
plt.title("M957 Treatment")
plt.xlabel("Timepoints")
plt.ylabel("Tumor Volume (mm3)")
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
weight=capomulin['Weight (g)'].tolist()
tumor_volume=capomulin['Tumor Volume (mm3)'].tolist()
plt.scatter(weight, tumor_volume, marker='o', facecolors='red')
plt.title("Weight/Tumor Volume for Capomulin")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
x_values = weight
y_values = tumor_volume
correlation=st.pearsonr(weight, tumor_volume)
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
for x in x_values:
    regress_values = x * slope + intercept

plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
```
